```python
# Convertir los ONNX FP32 a FP16
import onnx
from onnxconverter_common import float16
from pathlib import Path

EXPORTED_DIR = Path('/content/drive/MyDrive/UOC/TFM AI/xlab/outputs/exported')

for model_dir in EXPORTED_DIR.iterdir():
    if not model_dir.is_dir():
        continue
    fp32_path = model_dir / f'{model_dir.name}.onnx'
    fp16_path = model_dir / f'{model_dir.name}_fp16.onnx'

    if not fp32_path.exists():
        continue

    print(f'\n>>> {model_dir.name}')
    model = onnx.load(str(fp32_path))
    model_fp16 = float16.convert_float_to_float16(model, keep_io_types=True)
    onnx.save(model_fp16, str(fp16_path))

    size_fp32 = fp32_path.stat().st_size / 1024 / 1024
    size_fp16 = fp16_path.stat().st_size / 1024 / 1024
    print(f'  FP32: {size_fp32:.1f} MB')
    print(f'  FP16: {size_fp16:.1f} MB  ({size_fp16/size_fp32*100:.0f}%)')
```

```python
%%writefile pack_test_for_edge.py
import argparse, csv, hashlib, os, shutil, sys, tarfile
from pathlib import Path

def main():
    ap = argparse.ArgumentParser()
    ap.add_argument("--csv", required=True)
    ap.add_argument("--path-col", default="abs_path")
    ap.add_argument("--label-col", default="label")
    ap.add_argument("--root", default="")
    ap.add_argument("--out", default="edge_test_pkg")
    args = ap.parse_args()

    csv_path = Path(args.csv)
    if not csv_path.exists():
        sys.exit(f"ERROR: no existe el CSV: {csv_path}")

    out_dir = Path(args.out)
    img_dir = out_dir / "images"
    img_dir.mkdir(parents=True, exist_ok=True)

    with open(csv_path, newline="", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        if args.path_col not in reader.fieldnames:
            sys.exit(f"ERROR: columna '{args.path_col}' no existe. Hay: {reader.fieldnames}")
        rows = list(reader)
        has_class = "class_name" in reader.fieldnames
        has_source = "source" in reader.fieldnames

    print(f"Filas en test.csv: {len(rows)}")
    new_rows, missing, copied, seen = [], 0, 0, {}

    for row in rows:
        raw = row[args.path_col].strip()
        label = row[args.label_col].strip()
        class_name = row.get("class_name", "").strip() if has_class else ""
        source = row.get("source", "").strip() if has_source else ""

        src = Path(raw)
        if args.root and not src.is_absolute():
            src = Path(args.root) / raw

        if not src.exists():
            missing += 1
            if missing <= 10:
                print(f"  [FALTA] {src}")
            continue

        h = hashlib.md5(str(src).encode()).hexdigest()[:8]
        # Sanear nombre: quitar espacios y parentesis problematicos
        base = src.name.replace(" ", "_").replace("(", "").replace(")", "")
        dest_name = f"{label}__{h}__{base}"
        if dest_name in seen:
            dest_name = f"{label}__{h}_{copied}__{base}"
        seen[dest_name] = True

        shutil.copy2(src, img_dir / dest_name)
        copied += 1
        new_rows.append({
            "filepath": f"images/{dest_name}",
            "label": label, "class_name": class_name, "source": source,
        })

    new_csv = out_dir / "test_edge.csv"
    with open(new_csv, "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=["filepath", "label", "class_name", "source"])
        writer.writeheader()
        writer.writerows(new_rows)

    print(f"\nImagenes copiadas: {copied}")
    print(f"Imagenes faltantes: {missing}")
    print(f"CSV portable: {new_csv}")

    tar_path = Path(f"{args.out}.tar.gz")
    with tarfile.open(tar_path, "w:gz") as tar:
        tar.add(out_dir, arcname=out_dir.name)
    print(f"\nPaquete creado: {tar_path}  ({tar_path.stat().st_size/1e6:.1f} MB)")

if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
import os, sys
import numpy as np
import pandas as pd
from PIL import Image

#  CONFIG вЂ” EDITA 
CHECKPOINTS_DIR = CSV_IN  = "/content/drive/MyDrive/UOC/TFM AI/xlab/outputs/checkpoints"          # config.CHECKPOINTS_DIR del entrenamiento
TEST_CSV        = "/content/drive/MyDrive/UOC/TFM AI/xlab/outputs/corpus/test.csv"  # filepath,label,source
OUT_CSV         = "/content/drive/MyDrive/UOC/TFM AI/xlab/outputs/results/preds_pytorch_reference.csv"
IMG_SIZE        = 224
N_CLASSES       = 3
BATCH           = 32

MODELS = {
    "mobilenetv4_conv_small": "mobilenetv4_conv_small.e2400_r224_in1k",
    "efficientnetv2_b0":      "tf_efficientnetv2_b0.in1k",  
    "efficientformerv2_s0":   "efficientformerv2_s0.snap_dist_in1k",
    "repvit_m1":              "repvit_m1.dist_in1k",
}
IDX_TO_CLASS = {0: "healthy", 1: "early_blight", 2: "late_blight"}


try:
    import torch, timm
    import albumentations as A
    from albumentations.pytorch import ToTensorV2
except ImportError as e:
    sys.exit(f"[ERROR] Falta dependencia: {e}\n"
             f"  pip install torch timm albumentations opencv-python")

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
IMAGENET_MEAN = [0.485, 0.456, 0.406]
IMAGENET_STD  = [0.229, 0.224, 0.225]


def get_eval_transforms(img_size=IMG_SIZE):
    return A.Compose([
        A.Resize(int(img_size * 1.14), int(img_size * 1.14)),
        A.CenterCrop(img_size, img_size),
        A.Normalize(mean=IMAGENET_MEAN, std=IMAGENET_STD),
        ToTensorV2(),
    ])


def f1_macro(y_true, y_pred, n=N_CLASSES):
    y_true, y_pred = np.asarray(y_true), np.asarray(y_pred)
    f1s = []
    for c in range(n):
        tp = np.sum((y_pred == c) & (y_true == c))
        fp = np.sum((y_pred == c) & (y_true != c))
        fn = np.sum((y_pred != c) & (y_true == c))
        p = tp / (tp + fp) if (tp + fp) else 0.0
        r = tp / (tp + fn) if (tp + fn) else 0.0
        f1s.append(2 * p * r / (p + r) if (p + r) else 0.0)
    return float(np.mean(f1s))


def recall_per_class(y_true, y_pred, n=N_CLASSES):
    y_true, y_pred = np.asarray(y_true), np.asarray(y_pred)
    out = {}
    for c in range(n):
        fn_tp = np.sum(y_true == c)
        tp = np.sum((y_pred == c) & (y_true == c))
        out[IDX_TO_CLASS[c]] = float(tp / fn_tp) if fn_tp else float("nan")
    return out


def load_model(key, timm_name, ckpt_path):
    try:
        model = timm.create_model(timm_name, pretrained=False, num_classes=N_CLASSES)
    except Exception as e:
        cands = timm.list_models(f"*{timm_name.split('_')[0]}*")[:20]
        sys.exit(f"[ERROR] create_model('{timm_name}') fallo: {e}\n"
                 f"  Candidatos timm parecidos: {cands}")
    ckpt = torch.load(ckpt_path, map_location="cpu")
    state = ckpt
    for k in ("model_state_dict", "state_dict", "model"):
        if isinstance(ckpt, dict) and k in ckpt:
            state = ckpt[k]; break
    state = { (kk[7:] if kk.startswith("module.") else kk): vv for kk, vv in state.items() }
    miss, unexp = model.load_state_dict(state, strict=False)
    if miss or unexp:
        print(f"  [aviso] missing={len(miss)} unexpected={len(unexp)} "
              f"(ojo si la cabeza no encaja)")
    return model.eval().to(DEVICE)



df = pd.read_csv(TEST_CSV).reset_index(drop=True)
df["image_id"] = df.index
y_true = df["label"].astype(int).to_numpy()
if "source" not in df: df["source"] = "unknown"
tf = get_eval_transforms()
print(f"[OK] {len(df)} imagenes | device={DEVICE}\n")

rows, summary = [], []
for key, timm_name in MODELS.items():
    ckpt = os.path.join(CHECKPOINTS_DIR, f"{key}_best.pth")
    if not os.path.isfile(ckpt):
        print(f"[SKIP] {key}: no existe {ckpt}"); continue
    print(f"в–¶ {key}  (timm: {timm_name})")
    model = load_model(key, timm_name, ckpt)

    preds = np.empty(len(df), dtype=np.int64)
    with torch.no_grad():
        for i in range(0, len(df), BATCH):
            chunk = df.iloc[i:i + BATCH]
            xs = []
            for _, r in chunk.iterrows():
                img = np.array(Image.open(r["abs_path"]).convert("RGB"))
                xs.append(tf(image=img)["image"])
            x = torch.stack(xs).to(DEVICE)
            logits = model(x)
            preds[i:i + len(chunk)] = logits.argmax(1).cpu().numpy()

    f1 = f1_macro(y_true, preds)
    acc = float(np.mean(y_true == preds))
    rec = recall_per_class(y_true, preds)
    print(f"  в†’ accuracy={acc:.4f}  F1-macro={f1:.4f}")
    print(f"     recall/clase: " +
          "  ".join(f"{k}={v:.4f}" for k, v in rec.items()))
    for src in sorted(df["source"].unique()):
        m = (df["source"] == src).to_numpy()
        print(f"     {src:<28} acc={np.mean(y_true[m]==preds[m]):.4f} (n={int(m.sum())})")
    summary.append({"model": key, "accuracy": round(acc, 4),
                    "f1_macro": round(f1, 4), **{f"recall_{k}": round(v,4) for k,v in rec.items()}})
    for iid, src, yt, yp in zip(df["image_id"], df["source"], y_true, preds):
        rows.append({"image_id": int(iid), "source": src, "model": key,
                      "precision": "PYTORCH_REF", "y_true": int(yt), "y_pred": int(yp)})
    del model
    if DEVICE == "cuda": torch.cuda.empty_cache()
    print()

if not rows:
    sys.exit("[ERROR] Ningun checkpoint evaluado. Revisa CHECKPOINTS_DIR.")
os.makedirs(os.path.dirname(OUT_CSV) or ".", exist_ok=True)
pd.DataFrame(rows).to_csv(OUT_CSV, index=False)
print("=" * 64)
print(f"[OK] Escrito: {OUT_CSV}")
print(pd.DataFrame(summary).to_string(index=False))
```

```python
import sys, os
import numpy as np
import pandas as pd
import torch
from PIL import Image
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
from pytorch_grad_cam import GradCAMPlusPlus
from pytorch_grad_cam.utils.image import show_cam_on_image
from pytorch_grad_cam.utils.model_targets import ClassifierOutputTarget

PROJECT = "/content/edge-blight-classifier/project"
sys.path.append(PROJECT)
import config
from src.models import create_model
from src.transforms import get_eval_transforms

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Dispositivo:", device)

CKPT_DIR = config.CHECKPOINTS_DIR
print("Checkpoints:", os.listdir(CKPT_DIR))

BASE = "/content/drive/MyDrive/UOC/TFM AI"
OUT_DIR = f"{BASE}/xlab/outputs/figuras"
os.makedirs(OUT_DIR, exist_ok=True)

CSV = config.CORPUS_DIR / "test.csv"    
COL_PATH, COL_LABEL = "abs_path", "label"  
N_PER_CLASS = 2         
SEED = 428
IMG_SIZE = 224

SEARCH_DIRS = [f"{BASE}/xlab/dataset", f"{BASE}/dataset", "/content/dataset"]

CLASS_ES = ["Sano", "TizГіn temprano", "TizГіn tardГ­o"]
MODELS = list(config.MODEL_REGISTRY.keys())
SHORT = {"mobilenetv4_conv_small":"MobileNetV4-S","efficientnetv2_b0":"EfficientNetV2-B0",
         "efficientformerv2_s0":"EfficientFormerV2-S0","repvit_m1":"RepViT-M1"}

eval_tf = get_eval_transforms(IMG_SIZE)
rng = np.random.default_rng(SEED)

df = pd.read_csv(CSV)

def resolve_path(p):
    if os.path.exists(p): return p
    name = os.path.basename(p)
    for bc in SEARCH_DIRS:
        if os.path.isdir(bc):
            for root, _, files in os.walk(bc):
                if name in files: return os.path.join(root, name)
    return None

samples = [] 
for label in [0, 1, 2]:
    cand = df[df[COL_LABEL] == label]
    if len(cand) == 0:
        print(f"[AVISO] sin imagenes para clase {label}"); continue
    order = rng.permutation(len(cand))
    taken = 0
    for i in order:
        if taken >= N_PER_CLASS: break
        p = resolve_path(str(cand.iloc[i][COL_PATH]))
        if p:
            samples.append((label, p)); taken += 1
    if taken == 0:
        print(f"[AVISO] no se resolvio ninguna ruta para clase {label}")

print("Muestras elegidas:")
for lab, p in samples:
    print(f"  clase {lab} ({CLASS_ES[lab]}): {os.path.basename(p)}")

def load_image(path):
    img = np.array(Image.open(path).convert("RGB"))
    x = eval_tf(image=img)["image"]
    if not torch.is_tensor(x):
        x = torch.from_numpy(np.transpose(x, (2,0,1)))
    tensor = x.unsqueeze(0).float()
    rgb = np.array(Image.open(path).convert("RGB").resize((IMG_SIZE, IMG_SIZE)),
                   dtype=np.float32) / 255.0
    return tensor, rgb

def get_target_layers(model):
    if hasattr(model, "stages"):    return [model.stages[-1]]
    if hasattr(model, "blocks"):    return [model.blocks[-1]]
    if hasattr(model, "conv_head"): return [model.conv_head]
    last = None
    for mm in model.modules():
        if isinstance(mm, torch.nn.Conv2d): last = mm
    return [last]

def load_trained_demo(model_key):
    model = create_model(model_key, pretrained=False)
    ckpt = os.path.join(CKPT_DIR, f"{model_key}_best.pth")
    sd = torch.load(ckpt, map_location=device)
    sd = sd.get("model", sd.get("state_dict", sd)) if isinstance(sd, dict) else sd
    missing, unexpected = model.load_state_dict(sd, strict=False)
    print(f"  {model_key}: missing={len(missing)} unexpected={len(unexpected)}")
    return model.eval().to(device)

def load_trained(model_key):
    model = create_model(model_key, pretrained=False)
    ckpt = os.path.join(CKPT_DIR, f"{model_key}_best.pth")
    raw = torch.load(ckpt, map_location=device)
    # localizar state_dict
    sd = raw
    for cand in ["state_dict","model","model_state_dict","net"]:
        if isinstance(raw, dict) and cand in raw:
            sd = raw[cand]; break
    for pref in ["model.", "module.", "backbone."]:
        if all(k.startswith(pref) for k in sd.keys()):
            sd = {k[len(pref):]: v for k,v in sd.items()}
            print(f"  (quitado prefijo '{pref}')")
            break
    missing, unexpected = model.load_state_dict(sd, strict=False)
    print(f"  {model_key}: missing={len(missing)} unexpected={len(unexpected)}")
    assert len(missing) < 5, f"CARGA INCORRECTA en {model_key}"
ncols = len(samples)
fig, axes = plt.subplots(len(MODELS), ncols, figsize=(3.2*ncols, 3.2*len(MODELS)))
if ncols == 1: axes = axes[:, None]

for r, key in enumerate(MODELS):
    model = load_trained(key)
    cam = GradCAMPlusPlus(model=model, target_layers=get_target_layers(model))
    for c, (label, path) in enumerate(samples):
        tensor, rgb = load_image(path); tensor = tensor.to(device)
        gray = cam(input_tensor=tensor, targets=[ClassifierOutputTarget(label)])[0]
        vis = show_cam_on_image(rgb, gray, use_rgb=True)
        ax = axes[r, c]; ax.imshow(vis); ax.axis("off")
        if r == 0:
            ax.set_title(CLASS_ES[label], fontsize=11, fontweight="bold")
        if c == 0:
            ax.text(-0.08, 0.5, SHORT.get(key, key), rotation=90, va="center",
                    ha="right", transform=ax.transAxes, fontsize=11, fontweight="bold")
    print(f"[OK] {SHORT.get(key, key)}")

fig.suptitle("Mapas Grad-CAM++ por arquitectura y clase", fontsize=14, fontweight="bold", y=1.0)
fig.tight_layout(rect=[0.02, 0, 1, 0.98])
for ext in ("png","pdf"):
    fig.savefig(f"{OUT_DIR}/gradcam_grid_new.{ext}", dpi=300, bbox_inches="tight")
plt.close(fig)
print(f"\nGuardado: {OUT_DIR}/gradcam_grid_new.png y .pdf")
```

```python
import os, sys
import numpy as np
import pandas as pd
from PIL import Image
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt

PROJECT = "/content/edge-blight-classifier/project"
sys.path.append(PROJECT)
import config

CSV = config.CORPUS_DIR / "test.csv"     
BASE = "/content/drive/MyDrive/UOC/TFM AI"
OUT_DIR = f"{BASE}/xlab/outputs/figuras"
os.makedirs(OUT_DIR, exist_ok=True)

df = pd.read_csv(CSV)
print("Columnas:", list(df.columns))
print("Fuentes:", df['source'].unique() if 'source' in df.columns else "NO HAY columna source")
print("Filas:", len(df))
```

```python
# ===== Ajustes =====
COL_PATH  = "abs_path"    
COL_SRC   = "source"
COL_LABEL = "label"      

N_PER_SOURCE = 5          
THUMB = 256              
SEED = 423

SOURCES = ["plantvillage", "mendeley_ue", "tilahun", "quizhpe"]
SRC_LBL = {"plantvillage":"PlantVillage\n(laboratorio)",
           "mendeley_ue":"Mendeley UE\n(campo, Bangladesh)",
           "tilahun":"Tilahun\n(campo, EtiopГ­a)",
           "quizhpe":"Andina\n(campo, Ecuador)"}
CLASS_TAG = {0:"sano", 1:"t. temprano", 2:"t. tardГ­o"}

rng = np.random.default_rng(SEED)

def resolve_path(p):
    if os.path.exists(p): return p
    name = os.path.basename(p)
    for bc in [f"{BASE}/dataset", f"{BASE}/datos", "/content/dataset"]:
        if os.path.isdir(bc):
            for root, _, files in os.walk(bc):
                if name in files: return os.path.join(root, name)
    return None

def square_thumb(path, size=THUMB):
    """Recorte central cuadrado + resize uniforme. Todas las imagenes saldran iguales."""
    img = Image.open(path).convert("RGB")
    w, h = img.size
    s = min(w, h)
    left, top = (w - s)//2, (h - s)//2
    img = img.crop((left, top, left+s, top+s)).resize((size, size), Image.LANCZOS)
    return np.asarray(img)

fig, axes = plt.subplots(len(SOURCES), N_PER_SOURCE,
                         figsize=(2.1*N_PER_SOURCE, 2.1*len(SOURCES)))
if len(SOURCES) == 1: axes = axes[None, :]

for r, src in enumerate(SOURCES):
    sub = df[df[COL_SRC] == src]
    picks = []
    if len(sub) > 0:
        labels_avail = sorted(sub[COL_LABEL].unique())
        per = max(1, N_PER_SOURCE // max(1, len(labels_avail)))
        for lab in labels_avail:
            cand = sub[sub[COL_LABEL] == lab]
            k = min(per, len(cand))
            idx = rng.choice(len(cand), size=k, replace=False)
            picks += [cand.iloc[i] for i in idx]
        while len(picks) < N_PER_SOURCE and len(sub) >= N_PER_SOURCE:
            extra = sub.iloc[rng.integers(len(sub))]
            picks.append(extra)
        picks = picks[:N_PER_SOURCE]

    for c in range(N_PER_SOURCE):
        ax = axes[r, c]; ax.axis("off")
        if c < len(picks):
            row = picks[c]
            path = resolve_path(str(row[COL_PATH]))
            if path:
                ax.imshow(square_thumb(path))
                ax.set_title(CLASS_TAG.get(int(row[COL_LABEL]), ""),
                             fontsize=8, pad=2)
            else:
                ax.text(0.5,0.5,"(no encontrada)",ha="center",va="center",
                        fontsize=7,color="gray",transform=ax.transAxes)
        else:
            ax.text(0.5,0.5,"вЂ”",ha="center",va="center",
                    fontsize=10,color="lightgray",transform=ax.transAxes)
        if c == 0:
            ax.text(-0.12, 0.5, SRC_LBL[src], rotation=90, va="center", ha="right",
                    transform=ax.transAxes, fontsize=10, fontweight="bold")

fig.suptitle("Muestras representativas del corpus por fuente",
             fontsize=14, fontweight="bold", y=1.0)
fig.tight_layout(rect=[0.04, 0, 1, 0.98])
for ext in ("png","pdf"):
    fig.savefig(f"{OUT_DIR}/muestras_corpus.{ext}", dpi=300, bbox_inches="tight")
plt.close(fig)
print(f"Guardado: {OUT_DIR}/muestras_corpus.png y .pdf")
```
